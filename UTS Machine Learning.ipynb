1) Apa yang Dimaksud dengan Machine Learning
Machine Learning (ML) adalah cabang dari kecerdasan buatan (Artificial Intelligence, AI) yang berfokus pada pengembangan algoritma dan teknik yang memungkinkan komputer untuk belajar dari dan membuat prediksi atau keputusan berdasarkan data. Inti dari Machine Learning adalah penggunaan algoritma yang dapat memproses data input dan menghasilkan model yang dapat menggeneralisasi pola-pola dalam data tersebut. Dengan kata lain, komputer menggunakan data untuk "belajar" dan meningkatkan kinerjanya pada tugas tertentu tanpa memerlukan program eksplisit untuk setiap tugas.


 2) Contoh Penerapan Implementasi Machine Learning dalam Kehidupan Sehari-hari
Contoh 1: Penggunaan Asisten Virtual (Seperti Siri, Google Assistant, Alexa)
- Mengapa Membutuhkannya: Asisten virtual membantu pengguna dengan berbagai tugas seperti menjawab pertanyaan, mengatur pengingat, mengirim pesan, dan mengontrol perangkat rumah pintar.
- Manfaat:

Contoh 2: Rekomendasi Konten di Platform Streaming (Seperti Netflix, Spotify)

  - Retention: Dengan menyediakan konten yang relevan, pengguna lebih cenderung tetap menggunakan layanan tersebut.

 

3) Macam-macam Taxonomi dalam Penerapan Machine Learning
Machine Learning dapat dikategorikan ke dalam beberapa jenis berdasarkan bagaimana algoritma belajar dari data. Berikut adalah macam-macam taxonomi dalam penerapan Machine Learning:
 A. Supervised Learning
- Definisi: Algoritma dilatih menggunakan data yang sudah diberi label, yang berarti setiap data input sudah memiliki output yang benar.
- Contoh: 
  - Klasifikasi: Memprediksi kategori dari suatu data, seperti klasifikasi email menjadi spam atau tidak spam.
  - Regresi: Memprediksi nilai kontinu, seperti harga rumah berdasarkan fitur-fitur seperti lokasi, ukuran, dll.
- Manfaat: Sangat efektif untuk prediksi dan pengklasifikasian ketika data berlabel tersedia.

 B. Unsupervised Learning
- Definisi: Algoritma dilatih menggunakan data yang tidak diberi label. Algoritma mencoba menemukan struktur atau pola yang tersembunyi dalam data.
- Contoh:
  - Clustering: Mengelompokkan data ke dalam cluster berdasarkan kemiripan, seperti mengelompokkan pelanggan berdasarkan perilaku belanja.
  - Dimensionality Reduction: Mengurangi jumlah fitur dalam dataset, seperti Principal Component Analysis (PCA).
- Manfaat: Berguna untuk eksplorasi data dan penemuan pola yang tidak terstruktur.

 C. Semi-supervised Learning
- Definisi: Menggabungkan sejumlah kecil data berlabel dengan sejumlah besar data tidak berlabel selama pelatihan.
- Contoh: Algoritma yang digunakan untuk meningkatkan akurasi klasifikasi teks dengan memanfaatkan sejumlah besar data teks yang tidak berlabel.
- Manfaat: Efektif ketika pelabelan data sangat mahal atau memakan waktu, memanfaatkan sebagian besar data yang tidak berlabel.

 D. Reinforcement Learning
- Definisi: Algoritma belajar dengan berinteraksi dengan lingkungan dan menerima umpan balik dalam bentuk reward atau punishment. Tujuannya adalah memaksimalkan reward kumulatif.
- Contoh: 
  - Game AI: Algoritma yang belajar untuk memainkan game dengan strategi optimal, seperti AlphaGo.
  - Robotika: Robot yang belajar untuk melakukan tugas melalui trial and error.
- Manfaat: Cocok untuk situasi di mana keputusan berurutan perlu dibuat dan hasil dari tindakan hanya diketahui setelah beberapa langkah.

 E. Self-supervised Learning
- Definisi: Sebuah jenis supervised learning di mana data dilabeli secara otomatis oleh algoritma, sering kali dengan memanfaatkan sebagian data untuk memprediksi bagian lain.
- Contoh: 
  - Natural Language Processing (NLP): Menggunakan model language seperti GPT-3 yang mempelajari pola dalam teks untuk menghasilkan teks yang realistis.
- Manfaat: Mengatasi masalah keterbatasan data berlabel dengan memanfaatkan data yang tidak berlabel secara efisien.

import pandas as pd

data = {
    'Hari': ['Senin', 'Selasa', 'Rabu', 'Kamis', 'Jumat', 'Sabtu', 'Minggu'],
    'Datang': [2, 3, 4, 1, 1, 5, 2],
    'Biaya': [30000*2, 35000*3, 25000*4, 15000*1, 20000*1, 30000*5, 35000*2],
    'Mahasiswa': ['Ani', 'Budi', 'Jono', 'Lono', 'Joni', 'Ani', 'Budi']
}

dataset = pd.DataFrame(data)
rata_rata_datang = dataset['Datang'].mean()
biaya_tertinggi_hari = dataset.loc[dataset['Biaya'].idxmax(), 'Hari']
biaya_tertinggi = dataset['Biaya'].max()
hari_biaya_lebih_110000 = dataset[dataset['Biaya'] > 110000]['Hari'].tolist()
mahasiswa_terbanyak = dataset['Mahasiswa'].value_counts().idxmax()
datang_minggu = dataset.loc[dataset['Hari'] == 'Minggu', 'Mahasiswa'].values[0]
biaya_terendah = dataset['Biaya'].min()
frekuensi_tertinggi = dataset['Datang'].max()
frekuensi_terendah = dataset['Datang'].min()

print(f"a) {rata_rata_datang:.2f}")
print(f"b) {biaya_tertinggi_hari}")
print(f"c) {', '.join(hari_biaya_lebih_110000)}")
print(f"d) {mahasiswa_terbanyak}")
print(f"e) {datang_minggu}")
print(f"f) Tertinggi: {biaya_tertinggi}, Terendah: {biaya_terendah}")
print(f"g) {frekuensi_terendah}")

````python
import pandas as pd
import matplotlib.pyplot as plt


data = {
    'Fakultas': ['Bisnis', 'D3 Perhotelan', 'ICT', 'Ilmu Komunikasi', 'Seni dan Desain'],
    'Jumlah Mahasiswa': [260, 28, 284, 465, 735],
    'Akreditasi': ['A', 'A', 'B', 'A', 'A']
}


info_mahasiswa = pd.DataFrame(data)


print(info_mahasiswa)


fig, ax = plt.subplots()


bars = ax.bar(info_mahasiswa['Fakultas'], info_mahasiswa['Jumlah Mahasiswa'], color=['#ff9999','#66b3ff','#99ff99','#ffcc99','#c2c2f0'])


ax.set_title('Jumlah Mahasiswa per Fakultas')
ax.set_xlabel('Fakultas')
ax.set_ylabel('Jumlah Mahasiswa')


plt.xticks(rotation=45)


for bar in bars:
    yval = bar.get_height()
    ax.text(bar.get_x() + bar.get_width()/2.0, yval, int(yval), va='bottom') # va: vertical alignment

plt.tight_layout()
plt.show()
````
